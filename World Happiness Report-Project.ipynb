```python
data_columns=['Country','Region','Happiness Score','Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Generosity','Trust (Government Corruption)']
```

```python
data=dt[data_columns].copy()
```

```python
happy_df=data.rename({'Country':'Country_name','Region':'regional_indicator','Happiness Score':'happiness_score','Economy (GDP per Capita)':'logged_GDP_per_capital','Health (Life Expectancy)':'healthy_lige_expectancy','Freedom':'freedom_to_make_life_choice','Generosity':'generosity','Trust (Government Corruption)':'perceptions_of_corruption'},axis=1)
```

```python
happy_df.head()||
```

```python
happy_df.isnull().sum()
```

```python
#plot b/w happiness and GDP

plt.rcParams['figure.figsize']=(15,7)
plt.title('Plot Between Happiness Score and GDP')
sns.scatterplot(x=happy_df.happiness_score,y=happy_df.logged_GDP_per_capital,hue=happy_df.regional_indicator,s=200);


plt.legend(loc='upper left',fontsize=10)
plt.xlable('happiness_score')
plt.ylabel('logged_GDP_per_capital')
```

```python
gdt_region=happy_df.groupby('regional_indicator')['logged_GDP_per_capital'].sum()

gdt_region
```

```python
gdt_region.plot.pie(autopct='%1.1f%%')
plt.title('GDP by Region')
plt.ylabel(" ")
```

```python
# Total countries

total_country=happy_df.groupby('regional_indicator')[['Country_name']].count()
```

```python
total_country
```

```python
#correlation map

cor=happy_df.corr(method='pearson')
f,ax=plt.subplots(figsize=(10,5))
sns.heatmap(cor,mask=np.zeros_like(cor,dtype=np.bool),cmap='Blues',square=True,ax=ax)
```

```python
# corruption in regions

corruption=happy_df.groupby('regional_indicator')[['perceptions_of_corruption']].mean()

corruption
```

```python
plt.rcParams['figure.figsize']=(12,8)
plt.title('Perception of corrption in various Regions')
plt.xlabel('Regions',fontsize=15)
plt.ylabel('Corruption Index',fontsize=15)
plt.xticks(rotation=30,ha='right')
plt.bar(corruption.index,corruption.perceptions_of_corruption)
```

```python
top_10=happy_df.head(10)
bottom_10= happy_df.tail(10)
```

```python
fig, axes=plt.subplots(1,2, figsize=(16,6))
plt.tight_layout(pad=2)
xlabels=top_10.Country_name
axes[0].set_title('Top 10 Happinest Countrys life Exepctancy')
axes[0].set_xticklabels(xlabels,rotation=45,ha='right')
sns.barplot(x=top_10.Country_name,y=top_10.healthy_lige_expectancy,ax=axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_xlabel('Life expecctancy')


# fig, axes=plt.subplots(1,2, figsize=(16,6))
# plt.tight_layout(pad=2)
xlabels=bottom_10.Country_name
axes[1].set_title('Bottom 10 least happy Countrys life Exepctancy')
axes[1].set_xticklabels(xlabels,rotation=45,ha='right')
sns.barplot(x=bottom_10.Country_name,y=bottom_10.healthy_lige_expectancy,ax=axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_xlabel('Life expectancy')



```

```python
plt.rcParams['figure.figsize']=(15,7)
sns.scatterplot(x=happy_df.freedom_to_make_life_choice,y=happy_df.happiness_score,hue=happy_df.regional_indicator,s=200)
plt.legend(loc='upper left',fontsize='12')
plt.xlabel('freedom to make life choices')
plt.ylabel('happiness Score')
```

```python
country=happy_df.sort_values(by='perceptions_of_corruption').head(10)
plt.rcParams['figure.figsize']=(12,6)
plt.title('Countries with most preception of corruption')
plt.xlabel('Country',fontsize=13)
plt.ylabel('Corruption Index',fontsize=13)
plt.xticks(rotation=30,ha='right')
plt.bar(country.Country_name,country.perceptions_of_corruption)
```
