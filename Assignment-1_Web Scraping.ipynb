# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.


```python
#Import labries 
from  bs4 import BeautifulSoup
import requests
```

```python
#URL Connected 
page=requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")

page
```

```python

soup=BeautifulSoup(page.content)
soup
```

# movie Name

```python

scaped_movies=soup.find_all("td",class_="titleColumn")

scaped_movies
```

```python
movies=[]

for movie in scaped_movies:
    movie=movie.get_text().replace("\n","")
    movie=movie.strip(" ")
    movies.append(movie)
    
movies
```

# Rating

```python
# scrap rating for movies
scraped_rating=soup.find_all('td',class_='ratingColumn imdbRating')

scraped_rating
```

```python
ratings=[]

for rating in scraped_rating:
    rating=rating.get_text().replace("\n","")
    ratings.append(rating)
ratings
```

```python
##store the scaped data
```

```python
import pandas as pd
data=pd.DataFrame()
data['Movie Name']=movies
data['Ratings']=ratings
data.head(100)
```

```python
data.to_csv('imdb to movies.csv',index=False)
```
